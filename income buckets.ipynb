### Median income, 2000 v. 2014

Inflation adjustment using uscpi python library (I had to modify the library to collect the rest of the 2014 data, which is absent from the BLS source file.

```python
import pandas as pd
```

```python
# 2014 dollars
med_14 = pd.read_csv("data/ACS_14_5YR_S1901_with_ann.csv",skiprows=0)
med_00 = pd.read_csv("data/DEC_00_SF3_DP3_with_ann.csv",skiprows=0)
```

```python
hh_income_col_00 = "HC01_VC64"
hh_income_col_14 = "HC01_EST_VC13"
```

```python
median_income_2014 = med_14[["GEO.display-label",hh_income_col_14]]
```

```python
median_income_2000 = med_00[["GEO.display-label",hh_income_col_00]]
```

```python
median_income_2000
```

```python
from uscpi import UsCpi
cpi = UsCpi() # downloads the latest CPI data
```

```python
merged = median_income_2000.merge(median_income_2014,on="GEO.display-label")
merged.columns = ["town","2000","2014"]
merged["2000"] = pd.to_numeric(merged["2000"],errors="coerce")
merged["2014"] = pd.to_numeric(merged["2014"],errors="coerce")

merged["2000_adj"] = merged["2000"].apply(lambda x: cpi.value_with_inflation(x, 2000, 2014))
merged["diff"] = merged["2014"] - merged["2000_adj"] 
merged["diff_pct"] = merged["diff"] * 100 / merged["2000_adj"]
merged["beat"] = merged["diff"] > 0.0
```

```python

merged["beat"].value_counts()
```

```python
merged.to_csv("income_change.csv")
```
